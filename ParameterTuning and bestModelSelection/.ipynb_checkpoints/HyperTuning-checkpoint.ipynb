```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
# import data
digits = load_digits()
```

```python
dir(digits)
```

```python
data = pd.DataFrame(digits.data,columns = digits.feature_names)
```

```python
data['target'] = digits.target
```

```python
data.sample(1)
```

```python
# check and clean data
data.columns[data.isna().any()]
```

```python
# spliting data
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(data.drop('target',axis = 'columns'),data.target,test_size=30)
```

```python
# model selection

from sklearn.model_selection import GridSearchCV
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
model = {
    'svm' : {
        'model' : svm.SVC(),
        'para' :{
            'kernel': ['rbf','linear'] ,
            'C':[1,10,20]
        }
    },
    'RandomForestClassifier' : {
        'model' : RandomForestClassifier(),
        'para' : {
            'n_estimators' : [1,5,10]
        }
    },
    
    'LogisticRegression' :{
        'model' : LogisticRegression(),
        'para' : {
            'C': [1,5,10]
        }
    },
    
    'naive_bayes_gaussian' : {
        'model' : GaussianNB() ,
        'para' : { }
    },
    
    'naive_bayes_multi' : {
        'model' :  MultinomialNB(),
        'para' :{}
    },
    'decision_tree': {
        'model': DecisionTreeClassifier(),
        'para': {
            'criterion': ['gini','entropy'],       
        }
    }    
}
```

```python
# ?hyperTuning
cmf = []
for model_name,model_d in model.items():
    clf = GridSearchCV(model_d['model'],model_d['para'],cv = 5)
    clf.fit(x_train,y_train)
#     res = clf.cv_results_
    cmf.append({
        'model' : model_name,
        'best_score' : clf.best_score_,
        'best_para' : clf.best_params_
    })

df = pd.DataFrame(cmf)
```
